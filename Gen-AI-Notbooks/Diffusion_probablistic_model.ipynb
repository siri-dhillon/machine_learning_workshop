# Diffusion Probabilistic Models
The diffusers library, in conjunction with PyTorch, provides a powerful and convenient framework for implementing and utilizing diffusion probabilistic models for various tasks, including image generation. The following Python code demonstrates how to generate an image from a text prompt using a pre-trained Stable Diffusion model.

```python
from diffusers import StableDiffusionPipeline
from PIL import Image
```

```python
# Load the pre-trained Stable Diffusion model
pipeline = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
```

```python
# Set the device to CUDA if available, otherwise use CPU
# uncomment the following line and use if you have nvidia gpu
# pipeline.to("cuda" if pipeline.device.type == "cpu" else "cpu")

# if you have amd, please use this for this exercise and we will work with opencl at later point
pipeline.to("cpu")
```

```python
# Define the text prompt for image generation
prompt = "A futuristic cityscape at sunset, with flying cars and neon lights."

```

```python
# Generate the image
image = pipeline(prompt).images

```
